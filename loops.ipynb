```python
for i in range(1, 6):
    print(i)
```

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
sum = 0
for i in range(1, 11):
    sum += i

print("Sum:", sum)
```

```python
num = 5
for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```
