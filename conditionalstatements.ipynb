```python
age = 18
if age >= 18:
    print("You are eligible to vote")
```

```python
num = 10
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
else:
    print("Grade C")
```

```python
num = -5

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
a = 20
b = 30

if a > b:
    print("a is greater")
else:
    print("b is greater")
```
