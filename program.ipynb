```python
num = 1234
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

print("Reverse:", rev)
```

```python
num = 121
temp = num
rev = 0

while num > 0:
    rev = rev * 10 + num % 10
    num //= 10

if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
num = 153
temp = num
sum = 0

while num > 0:
    digit = num % 10
    sum += digit ** 3
    num //= 10

if temp == sum:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```
